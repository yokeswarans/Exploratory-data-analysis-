#### Data Cleaning

```python
df["Reviews"].unique()
```

```python
df["Reviews"].str.isnumeric().sum()
```

```python
df[~df["Reviews"].str.isnumeric()]
```

```python
df_copy=df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy["Reviews"].str.isnumeric()]
```

```python
## Convert Review Datatype to int
df_copy["Reviews"]=df_copy["Reviews"].astype(int)
```

```python
df_copy.info()
```

```python
df.head()
```

```python
df_copy['Size'].unique()
```

```python
## Convert the Millions into Thousands (M->K) and eliminate the k with empty
df_copy["Size"]=df_copy["Size"].str.replace('M','000')
df_copy["Size"]=df_copy["Size"].str.replace('k','')
df_copy['Size']=df_copy["Size"].replace('Varies with device',np.nan)
df_copy['Size']=df_copy["Size"].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
## cleaning the '+' in the installs col and '$' in the Price col
char_to_remove=['+','$',',']
cols_to_clean=['Installs','Price']
for item in char_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy["Price"].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype(int)
df_copy['Price']=df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
## Handle the Last update feature
df_copy['Last Updated'].unique()
```

```python
## give this column to the pandas inbuild func
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
#After cleaning save the dataset
df_copy.to_csv(r'D:\PracticeProjects\DataAnalysis\Data Cleaning with google playstore Dataset/Google Play Store Cleaned Dataset.csv')
```

```python
numeric_feature = [feature for feature in df_copy.columns if pd.api.types.is_numeric_dtype(df_copy[feature])]
categorical_feature = [feature for feature in df_copy.columns if pd.api.types.is_object_dtype(df_copy[feature])]

print("Numerical Feature : ",numeric_feature)
print("Categorical Feature: ",categorical_feature)
```

#### Featue Information

1. App: Name of the App.
2. Category: Category under which the App falls.
3. Rating: Apllications rating on playstore.
4. Reviews: Number of reviews of the App.
5. Size: Size of the App.
6. Install: Number of Installs of the App.
7. Type: If the App is Free or Paid.
8. Price: Price of the App (0 if it is free).
9. Content Rating: Appropriate Target Audience of the App.
10. Genres: Genre under which the App falls.
11. Last Updated: Date when the App was last updated.
12. Current Ver: Current version of the Application.
13. Android Ver: Minimum Android Version required to run the App



```python
## Proportion of count data on categorical columns
for cols in categorical_feature:
    print(df_copy[cols].value_counts(normalize=True)*100)
    print("*************************")
```

```python
## Proportion of Count data on numerical columns
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in  range(0,len(numeric_feature)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_copy[numeric_feature[i]],shade=True,color='r')
    plt.xlabel(numeric_feature[i])
    plt.tight_layout()
```

#### Observation

- Rating and year is left skewed while Reviews,Size,Install and Price are Right skewed

```python
## Categorical Columns
plt.figure(figsize=(20,15))
plt.suptitle('Univariate Analysis of Categorical Feature', fontsize=20,fontweight='bold',alpha=0.8,y=1.)

category=['Type','Content Rating']

for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df_copy[category[i]],palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
